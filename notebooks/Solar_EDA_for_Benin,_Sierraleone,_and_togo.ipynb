```python
# importing libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as zscore
```

```python
data_benin = pd.read_csv("../data/benin-malanville.csv")
data_sierraleones = pd.read_csv("../data/sierraleone-bumbuna.csv")
data_togo = pd.read_csv("../data/togo-dapaong_qc.csv")
```

```python
# reviewing head of benin data
data_benin.head()
```

```python
# reviewing data of sierraleon
data_sierraleones.head()
```

```python
# reviewing togo data
data_togo.head()
```

```python
print ("=============== benin data info ===============")
print(data_benin.info())
print("============== sierraleone data info ================")
print(data_sierraleones.info())
print("============== togo data info ================")
print(data_togo.info())
```

```python
print ("=============== benin data ===============")
data_benin.describe()
data_benin.isna().sum()

# clearly list columns with >5% missing data
threshold = len(data_benin) * 0.05
missing_cols_benin = data_benin.columns[data_benin.isna().sum() > threshold].tolist()
print(missing_cols_benin)

print ("=============== sierraleone data ===============")
data_sierraleones.describe()
data_sierraleones.isna().sum()

# clearly list col with >5% missing data
threshold = len(data_sierraleones) * 0.05
missing_cols_sierraleones = data_sierraleones.columns[data_sierraleones.isna().sum() > threshold].tolist
print(missing_cols_sierraleones)

print ("=============== togo data ===============")
data_togo.describe()
data_togo.isna().sum()

# clealy list col with >5% missing data
threshold = len(data_togo) * 0.05
missing_cols_togo = data_togo.columns[data_togo.isna().sum() > threshold].tolist
print(missing_cols_togo)
```

```python
# outlier detection and basic cleaning

cols_to_check = ['GHI', 'DNI', 'DHI', 'ModA', 'ModB', 'WS', 'WSgust']

data_benin_zscores = data_benin[cols_to_check].apply(zscore.zscore)
data_benin_outliers = (data_benin_zscores.abs() > 3).any(axis=1)

# Identify rows with outliers
data_benin_clean = data_benin[~data_benin_outliers].copy()

# adding impute for missing value
data_benin_clean.fillna(data_benin_clean.median(numeric_only=True), inplace=True)

# Dropping the 'Comments' column from the DataFrame
data_benin_clean.drop(columns=['Comments'], inplace=True)

print("================= data_benin-clean head======================")
data_benin_clean.head()
```

```python
data_sierraleones_zscores = data_sierraleones[cols_to_check].apply(zscore.zscore)
data_sierraleones_outliers = (data_sierraleones_zscores.abs() > 3).any(axis=1)

# Identify rows with outliers
data_sierraleones_clean = data_sierraleones[~data_sierraleones_outliers].copy()

# adding impute for missing value
data_sierraleones_clean.fillna(data_sierraleones_clean.median(numeric_only=True), inplace=True)

# Dropping the 'Comments' column from the DataFrame
data_sierraleones_clean.drop(columns=['Comments'], inplace=True)

print("================= data_sierraleon-clean head======================")
data_sierraleones_clean.head()
```

```python
data_togo_zscores = data_togo[cols_to_check].apply(zscore.zscore)
data_togo_outliers = (data_togo_zscores.abs() > 3).any(axis=1)

# Identify rows with outliers
data_togo_clean = data_togo[~data_togo_outliers].copy()

# adding impute for missing value
data_togo_clean.fillna(data_togo_clean.median(numeric_only=True), inplace=True)

# Dropping the 'Comments' column from the DataFrame
data_togo_clean.drop(columns=['Comments'], inplace=True)

print("================= data_togo-clean head======================")
data_togo_clean.head()
```

```python
# export cleaned data to my new folder cleaned
data_benin_clean.to_csv('../data/cleaned/benin-malanville-clean.csv', index=False)
data_sierraleones_clean.to_csv('../data/cleaned/sierraleone-bumbuna-clean.csv', index=False)
data_togo_clean.to_csv('../data/cleaned/togo-dapaong-clean.csv', index= False)
```

```python
# Visulalization of data
print("============================= benin data ============================")
data_benin_clean['Timestamp'] = pd.to_datetime(data_benin_clean['Timestamp'])
plt.figure(figsize=(12, 6))
plt.plot(data_benin_clean['Timestamp'], data_benin_clean[['GHI', 'DNI', 'DHI', 'Tamb']])
plt.title('GHI over Time Series in Benin')
plt.xlabel('Timestamp')
plt.ylabel('GHI')
plt.legend()
plt.show()

print("============================= sierraleone data ============================")
data_sierraleones_clean['Timestamp'] = pd.to_datetime(data_sierraleones_clean['Timestamp'])
plt.figure(figsize=(12, 6))
plt.plot(data_sierraleones_clean['Timestamp'], data_sierraleones_clean[['GHI', 'DNI', 'DHI', 'Tamb']])
plt.title('GHI over Time Series in Sierraleon')
plt.xlabel('Timestamp')
plt.ylabel('GHI')
plt.legend()
plt.show()

print("============================= togo data ============================")
data_togo_clean['Timestamp'] = pd.to_datetime(data_togo_clean['Timestamp'])
plt.figure(figsize=(12, 6))
plt.plot(data_togo_clean['Timestamp'], data_togo_clean[['GHI', 'DNI', 'DHI', 'Tamb']])
plt.title('GHI over Time Series in Togo')


plt.xlabel('Timestamp')
plt.ylabel('GHI')
plt.legend()
plt.show()
```

```python
# Extract month from Timestamp
print("============================= benin data ============================")
data_benin_clean['Month'] = data_benin_clean['Timestamp'].dt.month

# Aggregate monthly averages
monthly_avg_benin = data_benin_clean.groupby('Month')[['GHI', 'DNI', 'DHI', 'Tamb']].mean()

# Plot bar chart
monthly_avg_benin.plot(kind='bar', figsize=(12, 6))
plt.title('Monthly Average GHI, DNI, DHI, and Tamb in Benin')
plt.xlabel('Month')
plt.ylabel('Values')
plt.legend()
plt.show()


print("============================= sierraleone data ============================")
data_sierraleones_clean['Month'] = data_sierraleones_clean['Timestamp'].dt.month
# Aggregate monthly averages
monthly_avg_sierraleone = data_sierraleones_clean.groupby('Month')[['GHI', 'DNI', 'DHI', 'Tamb']].mean()
# Plot title
monthly_avg_sierraleone.plot(kind='bar', figsize=(12, 6))
plt.title('Monthly Average GHI, DNI, DHI, and Tamb in Sierraleone')
plt.xlabel('Month')
plt.ylabel('Values')
plt.legend()
plt.show()

print("============================= togo data ============================")
data_togo_clean['Month'] = data_togo_clean['Timestamp'].dt.month
# Aggregate monthly averages
monthly_avg_togo = data_sierraleones_clean.groupby('Month')[['GHI', 'DNI', 'DHI', 'Tamb']].mean()
# Plot title
monthly_avg_togo.plot(kind='bar', figsize=(12, 6))
plt.title('Monthly Average GHI, DNI, DHI, and Tamb in Togo')
plt.xlabel('Month')
plt.ylabel('Values')
plt.legend()
plt.show()
```

```python
# Cleaning impact analysis for Benin
print("============================= Benin Data ============================")
benin_clean_impact = data_benin_clean.groupby('Cleaning')[['ModA', 'ModB']].mean()
benin_clean_impact.plot(kind='bar', figsize=(8,6))
plt.title('Effect of Cleaning on ModA and ModB (Benin)')
plt.ylabel('Average Measurements (W/m)')
plt.show()

# Cleaning impact analysis for Sierra Leone
print("============================= Sierra Leone Data ============================")
sierraleone_clean_impact = data_sierraleones_clean.groupby('Cleaning')[['ModA', 'ModB']].mean()
sierraleone_clean_impact.plot(kind='bar', figsize=(8,6))
plt.title('Effect of Cleaning on ModA and ModB (Sierra Leone)')
plt.ylabel('Average Measurements (W/m)')
plt.show()

# Cleaning impact analysis for Togo
print("============================= Togo Data ============================")
togo_clean_impact = data_togo_clean.groupby('Cleaning')[['ModA', 'ModB']].mean()
togo_clean_impact.plot(kind='bar', figsize=(8,6))
plt.title('Effect of Cleaning on ModA and ModB (Togo)')
plt.ylabel('Average Measurements (W/m)')
plt.show()
```

```python
# Correlation Heatmap for Benin
print("============================= Benin Data ============================")
correlation_matrix_benin = data_benin_clean[['GHI', 'DNI', 'DHI', 'Tamb']].corr()

plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix_benin, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Heatmap of Solar Irradiance and Temperature (Benin)')
plt.show()

# Correlation Heatmap for Sierra Leone
print("============================= Sierra Leone Data ============================")
correlation_matrix_sierraleone = data_sierraleones_clean[['GHI', 'DNI', 'DHI', 'Tamb']].corr()

plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix_sierraleone, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Heatmap of Solar Irradiance and Temperature (Sierra Leone)')
plt.show()

# Correlation Heatmap for Togo
print("============================= Togo Data ============================")
correlation_matrix_togo = data_togo_clean[['GHI', 'DNI', 'DHI', 'Tamb']].corr()

plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix_togo, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Heatmap of Solar Irradiance and Temperature (Togo)')
plt.show()
```
